```python
def plot_all(df, target_col, num_cols=4):
    num_rows = math.ceil(len(df.columns)/num_cols)
    fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(16, 4*num_rows))
    sns.set(font_scale=1.2, style='whitegrid')
    for i, col_name in enumerate(df.columns):
        if col_name != target_col.name:
            ax = axes[i // num_cols, i % num_cols]
            if df[col_name].dtype == "O":
                sns.boxplot(x=df[col_name], y=target_col, ax=ax) #Categorical plot
            else:
                sns.regplot(x=target_col, y=df[col_name], ax=ax) #Numerical plot
            ax.set_xlabel(target_col.name, fontsize=10)
            ax.set_ylabel(col_name, fontsize=10)
            ax.tick_params(axis='both', which='major', labelsize=12)
    plt.tight_layout()
    plt.show()
```

```python
plot_all(data, data.Year,num_cols=2)
```

```python
# plot a bar graph for Year wise Per Capita Income
plt.figure(figsize=(22,15))
plot4=sns.barplot(data=data, x='Year',y='GDP Per Capita (in US $ PPP)', color='purple')
for bar in plot4.patches:
    plot4.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0,5),
                   textcoords='offset points')
plt.xlabel("Years")
plt.ylabel("Per Capita Income($)")
plt.title("Year wise Per Capita Income")
plt.savefig("Year-wise-PCI.jpg")
plt.show()

```
